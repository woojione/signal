```python
from scipy import signal
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import style


t = np.linspace(0,1.0,2001) #0~1.0 2001   2kHZ

sig_60Hz = np.sin(2*np.pi*60*t)    #60HZ
noise=np.random.rand(2001) #noise
sig_60Hz_noise = sig_60Hz+noise

style.use('dark_background')
f,plt_arr = plt.subplots(3, sharex=True)
f.suptitle('Signal Generation')

plt_arr[0].plot(sig_60Hz,color ='yellow')
plt_arr[0].set_title('60Hz signal', color ='yellow')

plt_arr[1].plot(noise,color ='cyan')
plt_arr[1].set_title('noise signal', color ='cyan')

plt_arr[2].plot(sig_60Hz_noise,color ='white')
plt_arr[2].set_title('60Hz_noise signal', color ='white')


plt.show()
```

```python
#
average_60Hz=np.mean(sig_60Hz)
average_noise=np.mean(noise)
average_60Hz_noise=np.mean(sig_60Hz_noise)
print('average_60Hz: ',average_60Hz)
print('average_noise: ',average_noise)
print('average_60Hz_noise: ',average_60Hz_noise)
```

```python
#
std_60Hz=np.std(sig_60Hz)
std_noise=np.std(noise)
std_60Hz_noise=np.std(sig_60Hz_noise)
print('std_60Hz: ', std_60Hz)
print('std_noise: ',std_noise)
print('std_60Hz_noise: ',std_60Hz_noise)
```

```python
from scipy.stats import norm 

# 

x = np.arange(-5, 5, 0.001) 
plt.figure(figsize=(15,10)) #(,)
plt.title('Gaussian') 
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid() #
plt.plot(x, norm.pdf(x, loc=average_60Hz_noise, scale=std_60Hz_noise)) #loc=, scale=
plt.show()
```
