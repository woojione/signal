```python
import librosa
import librosa.display
import scipy as sp
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = 1
t = np.arange(0,10,0.001)

sin = np.sin(2*np.pi * (f * t))
sin2 = np.sin(2*np.pi * (5*f * t))
sin3 = np.sin(2*np.pi * (10*f * t))

sum_signal = sin + sin2 + sin3

plt.figure(figsize=(15, 10))

plt.subplot(4, 1, 1) #4행 1열짜리 중 1번째
plt.plot(t, sum_signal, color="r")

plt.subplot(4, 1, 2)
plt.plot(t, sin, color='g')

plt.subplot(4, 1, 3)
plt.plot(t, sin2)

plt.subplot(4, 1, 4)
plt.plot(t, sin3, color='purple')

plt.show()
```

```python
n=len(sum_signal)
k=np.arange(n)
fs=1/0.001
T=n/fs
freq=k/T
freq=freq[range(int(n/2))]
```

```python
Y=np.fft.fft(sum_signal)/n
Y=Y[range(int(n/2))]
```

```python
fig, ax = plt.subplots(2, 1, figsize=(12,8))
ax[0].plot(t, sum_signal)
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Amplitude'); ax[0].grid(True)
ax[1].plot(freq, abs(Y), 'r', linestyle=' ', marker='^') 
ax[1].set_xlabel('Freq (Hz)')
ax[1].set_ylabel('|Y(freq)|')
ax[1].vlines(freq, [0], abs(Y))
ax[1].set_xlim([0, 20]); ax[1].grid(True)
plt.show()
```

```python
A=np.fft.fft(sin)/n
A=A[range(int(n/2))]

B=np.fft.fft(sin2)/n
B=B[range(int(n/2))]

C=np.fft.fft(sin3)/n
C=C[range(int(n/2))]
```

```python
fig, ax = plt.subplots(3, 1, figsize=(12,20))
ax[0].set_title('1kHz')
ax[0].plot(freq, abs(A), color='b', linestyle=' ', marker='^') 
ax[0].set_xlabel('Freq (Hz)')
ax[0].set_ylabel('|Y(freq)|')
ax[0].vlines(freq, [0], abs(A))
ax[0].set_xlim([0, 20]); ax[0].grid(True)

ax[1].set_title('5kHz')
ax[1].plot(freq, abs(B), color='g', linestyle=' ', marker='^') #기본 스타일 라인, ^마커
ax[1].set_xlabel('Freq (Hz)') #x축 라벨
ax[1].set_ylabel('|Y(freq)|') #y축 라벨
ax[1].vlines(freq, [0], abs(B)) #선
ax[1].set_xlim([0, 20]); ax[1].grid(True) #x축 범위, 격자

ax[2].set_title('10kHz')
ax[2].plot(freq, abs(C), color='r', linestyle=' ', marker='^') 
ax[2].set_xlabel('Freq (Hz)')
ax[2].set_ylabel('|Y(freq)|')
ax[2].vlines(freq,[0], abs(C))
ax[2].set_xlim([0, 20]); ax[2].grid(True)
plt.show()
```
